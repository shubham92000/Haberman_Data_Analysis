```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

HaberMan = pd.read_csv("haberman.csv")
```

```python
one = HaberMan.loc[HaberMan["status"] == 1]
```

```python
two = HaberMan.loc[HaberMan["status"] == 2]
```

```python
print(two.head())
```

```python
print("univariate analysis")
#pdf
```

```python
sns.FacetGrid(HaberMan, hue="status", height=5).map(sns.distplot,"nodes").add_legend();
plt.show();
```

```python
sns.FacetGrid(HaberMan, hue="status", height=5).map(sns.distplot, "age").add_legend();
plt.show();
```

```python
sns.FacetGrid(HaberMan, hue="status", height=5).map(sns.distplot, "year").add_legend();
plt.show();
```

```python
#cdf
```

```python
counts, bin_edges = np.histogram(one['nodes'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)

counts, bin_edges = np.histogram(two['nodes'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)
```

```python
counts, bin_edges = np.histogram(one['age'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)

counts, bin_edges = np.histogram(two['age'], bins=10,
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)
```

```python
counts, bin_edges = np.histogram(one['year'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)

counts, bin_edges = np.histogram(two['year'], bins=10,
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)
```

```python
print(np.mean(one['nodes']))
print(np.mean(two['nodes']))

print(np.std(one['nodes']))
print(np.std(two['nodes']))
```

```python
print(np.mean(one['age']))
print(np.mean(two['age']))

print(np.std(one['age']))
print(np.std(two['age']))
```

```python
#Median, Quantiles, MAD
print("\nMedians:")
print(np.median(one["nodes"]))
print(np.median(two["nodes"]))

print("\nQuantiles:")
print(np.percentile(one["nodes"],np.arange(0, 100, 25)))
print(np.percentile(two["nodes"],np.arange(0, 100, 25)))

print("\n90th Percentiles:")
print(np.percentile(one["nodes"],90))
print(np.percentile(two["nodes"],90))

from statsmodels import robust
print ("\nMedian Absolute Deviation")
print(robust.mad(one["nodes"]))
print(robust.mad(two["nodes"]))
```
